```python
from __future__ import print_function
import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import InputLayer, Conv2D, MaxPooling2D
from keras import backend as K
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
DF = pd.read_csv("data_final.csv",header=None)
X=DF.iloc[:,0:2]
Y=DF.iloc[:,2]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
print(X_train.shape)
np.random.random((1000,20)).shape
```

```python

model=Sequential()
model.add(Dense(40, input_dim=2, activation='relu'))
model.add(Dense(30, activation='sigmoid'))
model.add(Dense(25, activation='relu'))
model.add(Dense(20, activation='softmax'))
model.compile(loss='sparse_categorical_crossentropy', optimizer='rmsprop', metrics=['accuracy'])
#Y_train1= keras.utils.to_categorical(Y_train)
model.fit(X_train, Y_train, epochs=350, batch_size=10)
```

```python
accuracy = model.evaluate(X_train, Y_train)
print('Accuracy: %.2f' % (accuracy[1]*100))
```

```python
print(model.summary())
```

```python
accuracy = model.evaluate(X_test, Y_test)
print('Accuracy: %.2f' % (accuracy[1]*100))
```
